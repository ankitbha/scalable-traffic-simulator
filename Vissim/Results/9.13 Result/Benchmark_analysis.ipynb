```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
dl_no_net_Avg = df_no_net["DELAYAVG(ALL)"].groupby("TIMEINT").mean()
dl_eq_net_Avg = df_eq_net["DELAYAVG(ALL)"].groupby("TIMEINT").mean()
dl_scat_net_Avg = df_scat_net["DELAYAVG(ALL)"].groupby("TIMEINT").mean()
dl_scoot_net_Avg = df_scoot_net["DELAYAVG(ALL)"].groupby("TIMEINT").mean()
dl_ourdec_net_Avg = df_ourdec_net["DELAYAVG(ALL)"].groupby("TIMEINT").mean()

dl_no_net_Std = df_no_net["DELAYAVG(ALL)"].groupby("TIMEINT").std()
dl_eq_net_Std = df_eq_net["DELAYAVG(ALL)"].groupby("TIMEINT").std()
dl_scat_net_Std = df_scat_net["DELAYAVG(ALL)"].groupby("TIMEINT").std()
dl_scoot_net_Std = df_scoot_net["DELAYAVG(ALL)"].groupby("TIMEINT").std()
dl_ourdec_net_Std = df_ourdec_net["DELAYAVG(ALL)"].groupby("TIMEINT").std()

plt.errorbar(dl_no_net_Avg.index, dl_no_net_Avg, yerr=dl_no_net_Std, label="No Signaling Avg", fmt='-o')
plt.errorbar(dl_eq_net_Avg.index, dl_eq_net_Avg, yerr=dl_eq_net_Std, label="Equal Time Signaling Avg", fmt='-o')
plt.errorbar(dl_scat_net_Avg.index, dl_scat_net_Avg, yerr=dl_scat_net_Std, label="SCAT Avg", fmt='-o')
plt.errorbar(dl_scoot_net_Avg.index, dl_scoot_net_Avg, yerr=dl_scoot_net_Std, label="SCOOT Avg", fmt='-o')
plt.errorbar(dl_ourdec_net_Avg.index, dl_ourdec_net_Avg, yerr=dl_ourdec_net_Std, label="Self-regulation Protocol Avg", fmt='-o')


plt.legend(fontsize='small')
plt.xticks(rotation=90)  # Rotate labels by 90 degrees for better readability
plt.ylabel("Average Delay (in Seconds)")
plt.xlabel ("Signaling Cycles (in Seconds Interval)")
plt.title("Delay Curve on Single Lane Merge for 1:1 Load")
plt.show()
plt.close(fig)
```

```python
fig = plt.figure()
sp_no_net_Avg = df_no_net["SPEEDAVG(ALL)"].groupby("TIMEINT").mean()
sp_eq_net_Avg = df_eq_net["SPEEDAVG(ALL)"].groupby("TIMEINT").mean()
sp_scat_net_Avg = df_scat_net["SPEEDAVG(ALL)"].groupby("TIMEINT").mean()
sp_scoot_net_Avg = df_scoot_net["SPEEDAVG(ALL)"].groupby("TIMEINT").mean()
sp_ourdec_net_Avg = df_ourdec_net["SPEEDAVG(ALL)"].groupby("TIMEINT").mean()

sp_no_net_Std = df_no_net["SPEEDAVG(ALL)"].groupby("TIMEINT").std()
sp_eq_net_Std = df_eq_net["SPEEDAVG(ALL)"].groupby("TIMEINT").std()
sp_scat_net_Std = df_scat_net["SPEEDAVG(ALL)"].groupby("TIMEINT").std()
sp_scoot_net_Std = df_scoot_net["SPEEDAVG(ALL)"].groupby("TIMEINT").std()
sp_ourdec_net_Std = df_ourdec_net["SPEEDAVG(ALL)"].groupby("TIMEINT").std()

plt.errorbar(sp_no_net_Avg.index, sp_no_net_Avg, yerr=sp_no_net_Std, label="No Signaling Avg", fmt='-o')
plt.errorbar(sp_eq_net_Avg.index, sp_eq_net_Avg, yerr=sp_eq_net_Std, label="Equal Time Signaling Avg", fmt='-o')
plt.errorbar(sp_scat_net_Avg.index, sp_scat_net_Avg, yerr=sp_scat_net_Std, label="SCAT Avg", fmt='-o')
plt.errorbar(sp_scoot_net_Avg.index, sp_scoot_net_Avg, yerr=sp_scoot_net_Std, label="SCOOT Avg", fmt='-o')
plt.errorbar(sp_ourdec_net_Avg.index, sp_ourdec_net_Avg, yerr=sp_ourdec_net_Std, label="Self-regulation Protocol Avg", fmt='-o')

plt.legend()
plt.xticks(rotation=90)  # Rotate labels by 90 degrees for better readability
plt.ylabel("Average Speed (in Km/hr)")
plt.xlabel ("Signaling Cycles (in Seconds Interval)")
plt.title("Speed Curve on Single Lane Merge for 1:1 Load")
plt.show()
plt.close(fig)
```

```python
no_net_Arr = df_no_net["VEHARR(ALL)"].groupby("TIMEINT").mean()
eq_net_Arr = df_eq_net["VEHARR(ALL)"].groupby("TIMEINT").mean()
scat_net_Arr = df_scat_net["VEHARR(ALL)"].groupby("TIMEINT").mean()
scoot_net_Arr = df_scoot_net["VEHARR(ALL)"].groupby("TIMEINT").mean()
ourdec_net_Arr = df_ourdec_net["VEHARR(ALL)"].groupby("TIMEINT").mean()

no_net_Arr_std = df_no_net["VEHARR(ALL)"].groupby("TIMEINT").std()
eq_net_Arr_std = df_eq_net["VEHARR(ALL)"].groupby("TIMEINT").std()
scat_net_Arr_std = df_scat_net["VEHARR(ALL)"].groupby("TIMEINT").std()
scoot_net_Arr_std = df_scoot_net["VEHARR(ALL)"].groupby("TIMEINT").std()
ourdec_net_Arr_std = df_ourdec_net["VEHARR(ALL)"].groupby("TIMEINT").std()

thpt = [no_net_Arr[-1], eq_net_Arr[-1], scat_net_Arr[-1], scoot_net_Arr[-1], ourdec_net_Arr[-1]]
thpt_std = [no_net_Arr_std[-1], eq_net_Arr_std[-1], scat_net_Arr_std[-1], scoot_net_Arr_std[-1], ourdec_net_Arr_std[-1]]

stgy = ["No Signal", "Equal Signal", "SCAT", "SCOOT", "Self-regulation"]

fig = plt.figure()
plt.bar(stgy, thpt, yerr=thpt_std, capsize=5)
plt.xlabel("Control Strategy")
plt.ylabel("Average Throughput (in Number of Cars)")
plt.title("Throughput plot of different strategies for 1:1 load")
plt.show()
plt.close(fig)
```

```python
Cap = pd.read_csv("9.13 MAX CAP.csv", header=0, index_col=0)
```

```python
sim_end_time = Cap.mean(axis=1)
sim_end_time_std = Cap.std(axis=1)
fig = plt.figure()
Cap_name = [" Self-regulation", "No Signal", "Equal Signal", "SCOOT", "SCAT"]

fig, ax = plt.subplots()
ax.bar(Cap_name, sim_end_time, yerr=sim_end_time_std, capsize=5)  # capsize specifies the width of the caps on the error bars
ax.set_xlabel("Control Strategy")
ax.set_ylabel("Average Jam Time (in seconds)")
ax.set_title("Time to hit a jam in different strategies for 1:1 load")
plt.show()

```
