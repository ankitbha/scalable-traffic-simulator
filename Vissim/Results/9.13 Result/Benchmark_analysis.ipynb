```python
arr_no_net = df_no_net[df_no_net.index == "SUM"]
arr_eq_net = df_eq_net[df_eq_net.index == "SUM"]
arr_scat_net = df_scat_net[df_scat_net.index == "SUM"]
arr_scoot_net = df_scoot_net[df_scoot_net.index == "SUM"]
arr_ourdec_net = df_ourdec_net[df_ourdec_net.index == "SUM"]

no_net_Arr = arr_no_net["VEHARR(ALL)"].groupby("TIMEINT").mean()
eq_net_Arr = arr_eq_net["VEHARR(ALL)"].groupby("TIMEINT").mean()
scat_net_Arr = arr_scat_net["VEHARR(ALL)"].groupby("TIMEINT").mean()
scoot_net_Arr = arr_scoot_net["VEHARR(ALL)"].groupby("TIMEINT").mean()
ourdec_net_Arr = arr_ourdec_net["VEHARR(ALL)"].groupby("TIMEINT").mean()

no_net_Arr_std = arr_no_net["VEHARR(ALL)"].groupby("TIMEINT").std()
eq_net_Arr_std = arr_eq_net["VEHARR(ALL)"].groupby("TIMEINT").std()
scat_net_Arr_std = arr_scat_net["VEHARR(ALL)"].groupby("TIMEINT").std()
scoot_net_Arr_std = arr_scoot_net["VEHARR(ALL)"].groupby("TIMEINT").std()
ourdec_net_Arr_std = arr_ourdec_net["VEHARR(ALL)"].groupby("TIMEINT").std()

thpt = [no_net_Arr[-1], eq_net_Arr[-1], scat_net_Arr[-1], scoot_net_Arr[-1], ourdec_net_Arr[-1]]
thpt_std = [no_net_Arr_std[-1], eq_net_Arr_std[-1], scat_net_Arr_std[-1], scoot_net_Arr_std[-1], ourdec_net_Arr_std[-1]]

stgy = ["No Signal", "Equal Signal", "SCAT", "SCOOT", "Self-regulation"]

fig = plt.figure()
plt.bar(stgy, thpt, yerr=thpt_std, capsize=5)
plt.xlabel("Control Strategy")
plt.ylabel("Average Throughput (in Number of Cars)")
plt.title("Throughput plot of different strategies for 1:1 load")
plt.show()
plt.close(fig)
```

```python
Cap = pd.read_csv("9.13 MAX CAP.csv", header=0, index_col=0)
```

```python
sim_end_time = Cap.mean(axis=1)
sim_end_time_std = Cap.std(axis=1)
fig = plt.figure()
Cap_name = [" Self-regulation", "No Signal", "Equal Signal", "SCOOT", "SCAT"]

fig, ax = plt.subplots()
ax.bar(Cap_name, sim_end_time, yerr=sim_end_time_std, capsize=5)  # capsize specifies the width of the caps on the error bars
ax.set_xlabel("Control Strategy")
ax.set_ylabel("Average Jam Time (in seconds)")
ax.set_title("Time to hit a jam in different strategies for 1:1 load")
plt.show()

```
