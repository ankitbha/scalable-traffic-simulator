```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_no_net = pd.read_csv("MtoNj_no.csv", header=1, index_col=0)
```

```python
df_no_net
```

```python
df_eq_net = pd.read_csv("MtoNj_equal.csv", header=1, index_col=0)
```

```python
df_eq_net
```

```python
df_scat_net = pd.read_csv("MtoNj_scats.csv", header=1, index_col=0)
```

```python
df_scat_net
```

```python
df_scoot_net = pd.read_csv("MtoNj_scoot.csv", header=1, index_col=0)
```

```python
df_scoot_net
```

```python
df_ourdec_net = pd.read_csv("MtoNj_our.csv", header=1, index_col=0)
```

```python
df_ourdec_net
```

```python
fig = plt.figure()
dl_no_net_Avg = df_no_net["DELAYAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
dl_eq_net_Avg = df_eq_net["DELAYAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
dl_scat_net_Avg = df_scat_net["DELAYAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
dl_scoot_net_Avg = df_scoot_net["DELAYAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
dl_ourdec_net_Avg = df_ourdec_net["DELAYAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
plt.plot(dl_no_net_Avg[:len(dl_no_net_Avg)-1], label="No Signaling Avg")
plt.plot(dl_eq_net_Avg[:len(dl_eq_net_Avg)-1], label="Equal Time Signaling Avg")
plt.plot(dl_scat_net_Avg[:len(dl_scat_net_Avg)-1], label="SCAT Avg")
plt.plot(dl_scoot_net_Avg[:len(dl_scoot_net_Avg)-1], label="SCOOT Avg")
plt.plot(dl_ourdec_net_Avg[:len(dl_ourdec_net_Avg)-1], label="Our Decentralized Protocol Avg")
plt.legend()
plt.ylabel("Average Delay (in Seconds)")
plt.xlabel ("Signaling Cycles (in Seconds Interval)")
plt.title("Delay Curve on Single Lane Merge for 1:1 Load")
plt.show()
plt.close(fig)
```

```python
fig = plt.figure()
sp_no_net_Avg = df_no_net["SPEEDAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
sp_eq_net_Avg = df_eq_net["SPEEDAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
sp_scat_net_Avg = df_scat_net["SPEEDAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
sp_scoot_net_Avg = df_scoot_net["SPEEDAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
sp_ourdec_net_Avg = df_ourdec_net["SPEEDAVG(ALL)"].groupby("TIMEINT", sort=False).mean()
plt.plot(sp_no_net_Avg[:len(sp_no_net_Avg)-1], label="No Signaling Avg")
plt.plot(sp_eq_net_Avg[:len(sp_eq_net_Avg)-1], label="Equal Time Signaling Avg")
plt.plot(sp_scat_net_Avg[:len(sp_scat_net_Avg)-1], label="SCAT Avg")
plt.plot(sp_scoot_net_Avg[:len(sp_scoot_net_Avg)-1], label="SCOOT Avg")
plt.plot(sp_ourdec_net_Avg[:len(sp_ourdec_net_Avg)-1], label="Our Decentralized Protocol Avg")
plt.legend()
plt.ylabel("Average Speed (in Km/hr)")
plt.xlabel ("Signaling Cycles (in Seconds Interval)")
plt.title("Speed Curve on Single Lane Merge for 1:1 Load")
plt.show()
plt.close(fig)
```

```python
no_net_Arr = df_no_net["VEHARR(ALL)"].groupby("TIMEINT").sum()
eq_net_Arr = df_eq_net["VEHARR(ALL)"].groupby("TIMEINT").sum()
scat_net_Arr = df_scat_net["VEHARR(ALL)"].groupby("TIMEINT").sum()
scoot_net_Arr = df_scoot_net["VEHARR(ALL)"].groupby("TIMEINT").sum()
ourdec_net_Arr = df_ourdec_net["VEHARR(ALL)"].groupby("TIMEINT").sum()
thpt = []
thpt.append(no_net_Arr[len(no_net_Arr)-1])
thpt.append(eq_net_Arr[len(eq_net_Arr)-1])
thpt.append(scat_net_Arr[len(scat_net_Arr)-1])
thpt.append(scoot_net_Arr[len(scoot_net_Arr)-1])
thpt.append(ourdec_net_Arr[len(ourdec_net_Arr)-1])
stgy = ["No Signal", "Equal Signal", "SCAT", "SCOOT", "Decentralized"]

fig = plt.figure()
plt.bar(stgy, thpt)
plt.xlabel("Control Strategy")
plt.ylabel("Total Throughput (in Number of Cars)")
plt.title("Throughput plot of different strategies for 1:1 load")
plt.show()
plt.close(fig)
```

```python
Cap = pd.read_csv("MtoNj MAX CAP.csv", header=0, index_col=0)
```
